# KMeans Clustering 

```python
# Import necessary libraries
from sklearn.cluster import KMeans
 
# Define colors for the clusters
COLORS = ['b', 'r','y']
 
# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df_cleaned)
 
# Visualize the clusters
plt.figure(figsize=(15, 6))
 
plt.subplot(1, 2, 1)
plt.scatter(df_cleaned['sepal_length'], df_cleaned['sepal_width'], c=[COLORS[i] for i in kmeans.labels_])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('KMeans Clustering (Sepal)')
 
plt.subplot(1, 2, 2)
plt.scatter(df_cleaned['petal_length'], df_cleaned['petal_width'], c=[COLORS[i] for i in kmeans.labels_])
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('KMeans Clustering (Petal)')
 
plt.tight_layout()
plt.show()
```

```python
# within cluster sum of sqaures
wcs = []  
for i in range(1,10):
    kmeans = KMeans(n_clusters=i, init = 'k-means++',random_state=42)
    kmeans.fit(df_cleaned)
    wcs.append(kmeans.inertia_)
 
plt.plot(range(1,10),wcs)
plt.title('The elbow method')
plt.xlabel('No of clusters')
 
plt.ylabel('sum of sqaured distances')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sil_score = []
k_rng = range(2,11)
for i in k_rng:
    kmeans = KMeans(n_clusters=i)
    km_pred = kmeans.fit_predict(df_cleaned)
    sil_score.append(silhouette_score(df_cleaned, km_pred))
sil_score
```
